```python
import requests
import json

SERVER_URL = f'http://localhost:50021/'
SUCCESSFUL_RESPONCE = 200
VALIDATION_ERROR = 422

def audio_query(text, speaker=1):
    query_payload = {'text': text, 'speaker': speaker}
    query_response = requests.post(f'{SERVER_URL}audio_query', params=query_payload)
    print(query_response.status_code)
    if query_response.status_code != SUCCESSFUL_RESPONCE:
        return None
    
    query = query_response.json()
    return query

if __name__ == "__main__":
    # 読み上げたいテキスト
    text = "ダイナマイト"
    query_json = audio_query(text, 1)
    print(json.dumps(query_json, indent=2, ensure_ascii=False))

    # 音声合成の実行
    # synthesize_voice(text, speaker=1, filename="voicevox_output.wav")
```
